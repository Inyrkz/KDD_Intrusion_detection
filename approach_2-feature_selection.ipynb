```python
# create categories
dos = ['apache2', 'back', 'land', 'neptune', 'mailbomb',    
       'pod', 'processtable','smurf','teardrop','udpstorm','worm']

probe = ['ipsweep', 'mscan', 'nmap', 'portsweep', 'saint', 'satan']

u2r_r2l = ['buffer_overflow', 'loadmdoule', 'perl', 'ps', 'rootkit', 'sqlattack', 'xterm', 
           'ftp_write', 'guess_passwd', 'http_tunnel', 'imap', 'multihop', 'named', 'phf',
           'sendmail', 'snmpgetattack', 'snmpguess', 'spy', 'warezclient', 'warezmaster',
           'xclock','xsnoop']


# df['new_label'] = df.labels != 'normal'
def map_attack(attack):
    '''Function to map attack type to the four categories listed above'''
    if attack in dos:
        value = 'DoS'
    elif attack in probe:
        value = 'Probe'
    elif attack in u2r_r2l:
        value = 'U2R/R2L'
    else:
        value = 'Normal'
    return value

# Feature engineer a new column based on the map_attack() function
# for the training set and the test set
df['new_label'] = df.labels.map(map_attack)
df_test['new_label'] = df_test.labels.map(map_attack)
```

### Bivariate Analysis

```python
def viobox(df, x, y, suptitle):
    '''
    Function to compare the classes of a categorical variable with a numerical variable
    using a facetted boxplot and violin plot.
    
    df: DataFrame, Dataset
    x: str, categorical variable on the x-axis
    y: str, numeric variable on the y-axis
    suptitle: str, title of the plot
    color: str, color of the plot [default is base_color]
    '''
    base_color = sns.color_palette()[0]
    plt.figure(figsize=[20,5])
    plt.suptitle(suptitle, fontsize=15)
    plt.subplot(1,2,1)
    sns.boxplot(data=df, x=x, y=y, color=base_color);
    plt.subplot(1,2,2)
    sns.violinplot(data=df, x=x, y=y, color=base_color);
```

# Model Training

Define helper functions to evaluate the models and plot confusion matrix

```python
# Define Evaluation Metric Functions
from sklearn.metrics import confusion_matrix, accuracy_score, recall_score
from sklearn.metrics import precision_score, f1_score, ConfusionMatrixDisplay
from sklearn.metrics import classification_report


def evaluate_model(y_val, y_pred):
    """Function to evaluate model and return the metric of the model
  
    It returns a dictionary with the classification metrics.
    """
    accuracy = accuracy_score(y_val, y_pred)
    precision = precision_score(y_val, y_pred, average='weighted')  
    recall = recall_score(y_val, y_pred, average='weighted')
    f1 = f1_score(y_val, y_pred, average='weighted')
    result = {"accuracy_score": accuracy,
              "precision_score": precision,
              "recall_score": recall,
              "f1_score": f1}
    return result


def plot_confusion_matrix(y_val, y_pred, label):
    '''function to plot confusion matrix
    
    Args
    y_val: array. The validation set of the target variable.
    y_pred: array. Model's prediction.
    label: list. A list containing all the classes in the target variable

    Returns
    It returns a plot of the confusion matrix
    '''
    cm = confusion_matrix(y_val, y_pred)
    fig, ax = plt.subplots(figsize=(10,10))
    ConfusionMatrixDisplay(cm, display_labels=label).plot(ax=ax, values_format='', xticks_rotation='vertical')
    plt.show()


# class labels
label = ['DoS', 'Normal', 'Probe', 'U2R/R2L']
```

### Phase 1: Training with all features
Define a single function to train a Decision Tree, Random Forest, KNN & XGBoost algorithms

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier



RANDOM_STATE = 1

def train_models(X_train, X_test, y_train, y_test, label=label):
    '''Function to train Decision Tree, Random Forest, KNN & XGBoost algorithms.'''
    
    # Train the KNN algorithm
    print("Training the KNN algorithm...")
    knn = KNeighborsClassifier(n_neighbors=5, metric='minkowski', p=2)
    knn.fit(X_train, y_train)
    # Evaluate model on the training set
    print("Training Result - KNN")
    print(evaluate_model(y_train, knn.predict(X_train)))
    # Evaluate model on the test set
    y_pred = knn.predict(X_test)
    knn_result = evaluate_model(y_test, y_pred)
    print(knn_result)
    plot_confusion_matrix(y_test, y_pred, label)
    print(classification_report(y_test, y_pred)) 
    
    print("----------------------------------------------")
    print("\n\nTraining the decision tree algorithm...")
    
    # Train the Decision Tree algorithm
    dt = DecisionTreeClassifier(criterion="entropy",
                                random_state=RANDOM_STATE)
    dt.fit(X_train, y_train)
    # Evaluate model on the training set
    print("Training Result - Decision Tree")
    print(evaluate_model(y_train, dt.predict(X_train)))
    # Evaluate model on the test set
    y_pred = dt.predict(X_test)
    dt_result = evaluate_model(y_test, y_pred)
    print(dt_result)
    plot_confusion_matrix(y_test, y_pred, label)
    print(classification_report(y_test, y_pred)) 
    
    print("----------------------------------------------")
    print("\n\nTraining the random forest algorithm...")
    
    # Train the Random Forest algorithm
    rf = RandomForestClassifier(n_estimators=100,
                                criterion="entropy",
                                min_samples_split=25,
                                random_state=RANDOM_STATE)
    rf.fit(X_train, y_train)
    # Evaluate model on the training set
    print("Training Result - Random Forest")
    print(evaluate_model(y_train, rf.predict(X_train)))
    # Evaluate model on the test set
    y_pred = rf.predict(X_test)
    rf_result = evaluate_model(y_test, y_pred)
    print(rf_result)
    plot_confusion_matrix(y_test, y_pred, label)
    print(classification_report(y_test, y_pred)) 
    
    print("----------------------------------------------")
    print("\n\nTraining the XGBoost algorithm...")
    
    
    # Train the XGBoost algorithm
    xgb = XGBClassifier(n_estimators=100,
                        random_state=RANDOM_STATE,
                        eval_metric='mlogloss',
                        use_label_encoder=False)
    xgb.fit(X_train, y_train)
    # Evaluate model on the training set
    print("Training Result - XGBoost")
    print(evaluate_model(y_train, xgb.predict(X_train)))
    # Evaluate model on the test set
    y_pred = xgb.predict(X_test)
    xgb_result = evaluate_model(y_test, y_pred)
    print(xgb_result)
    plot_confusion_matrix(y_test, y_pred, label)
    print(classification_report(y_test, y_pred)) 
    
    print("----------------------------------------------")
    print("\n\nTraining Completed.")
  
```

```python
# Train models
train_models(scaled_X_train, 
             scaled_X_test, 
             encoded_y_train, 
             encoded_y_test, 
             label=label)
```
